# Import the required library

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

# 1. 

## Part A.

```python
data1 = pd.read_csv('/content/drive/MyDrive/Projects/AI_Assignments/Assignment_Practice_1/chocolates-univariate-csv.csv')
data1.drop(['Unnamed: 1'],axis=1)
X = np.array(data1['UGive']).reshape(len(data1),1)
y = np.array(data1['IGive'])
print('X shape : ',X.shape)
print('y shape : ',y.shape)
```

```python
M1 = LinearRegression()
M1.fit(X, y)
w0 = M1.intercept_
w1 = M1.coef_[0]
print('w0 : ',w0)
print('w1 : ',w1)
```

## Part B.

```python
data2 = pd.read_csv('/content/drive/MyDrive/Projects/AI_Assignments/Assignment_Practice_1/chocolates-multivariate-csv.csv')
X = np.array(data2[['UGive','Age']])
y = np.array(data2['IGive'])
print('X shape : ',X.shape)
print('y shape : ',y.shape)

```

```python
data3 = pd.read_csv('/content/drive/MyDrive/Projects/AI_Assignments/Assignment_Practice_1/iris.csv')
data3
X = np.array(data3[['sepal_length','sepal_width','petal_length','petal_width']])
y = np.array(data3['species'])
for i in range(len(y)):
  if y[i]=='setosa':
    y[i] = 1
  else :
    y[i] = 0  

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.1, random_state=100)

print('X shape : ',X.shape)
print('y shape : ',y.shape)
print('X_train : ',X_train.shape)
print('X_test : ',X_test.shape)
```

```python
M3 =  MLPClassifier(
    hidden_layer_sizes=(32),
    max_iter=100,
    solver="sgd",
    verbose=10,
    random_state=10,
    learning_rate_init=0.01)
M3.fit(X_train,list(y_train))

print('\n\nScore on test data : ',M3.score(X_test,list(y_test)),'\n\n')
```

```python
w = M3.coefs_
print('w : \n',w)
```
