# NL interface functions

```python
def PrintChoice(cho):
  print('You can opt these electives : ')
  for i in range(len(cho)):
    print('\t'+str(i+1)+'. '+cho[i])

def GetElective(opt):

  if opt=='alg':
    PrintChoice(['Randomized Algorithms','Introduction to Graduate Algorithms','Approximation Algorithms',
           'Advanced Algorithms','Modern Algorithm Design'])
  elif opt=='ssh':
    PrintChoice(['Philosophy of Technology','Urban Space and Political Power','Comparative Politics in Digital Era',
           'Technology and the Future of Work','Advanced Sociological Theory','Advance Ethnographic Research Methods'])  
  elif opt=='subbio':
    PrintChoice(['Foundations of Biology','Cell Biology and Bio-Chemistry','Cell Biology and Bio-Chemistry',
           'Foundations of Modern Biology','Introduction to Mathematical Biology','Introduction to Computational Neuroscience'])
  elif opt=='mathematics':
    PrintChoice(['Linear Optimization','Advanced Linear Algebra','Algebraic Number Theory','Variational Calculus and their Applications',
           'Convex Optimization','Combinatorial Optimization'])
  elif opt=='robot':
    PrintChoice(['Probability and Random Process','Reinforcement Learning','Robotics','Social Robotics'])
  elif opt=='dl':
    PrintChoice(['Advance Machine Learning','Deep Learning','Probabilitic Graphical Model','Bayesian Machine Learning'])
  elif opt=='beginmlyesbioyes':
    PrintChoice(['Machine Learning','Artifical Intelligence' ,'Data Mining','Data Science','Data Warehouse','Data Lifecycle Management',
           'Algorithms in Bioinformatics','Data Sciences for Genomics','Machine Learning for Biomedical Applications',
           'Biostatistics','Advanced Biometrics','Computing for Medicine'])
  elif opt=='beginmlyesbiono':
    PrintChoice(['Machine Learning','Artifical Intelligence' ,'Data Mining','Data Science','Data Warehouse',
           'Data Lifecycle Management'])
  elif opt=='beginmlno':
    PrintChoice(['Probability and Random Process','Probability and Statistics','Measure and Probability Theory',
           'Machine Learning','Data Mining'])
  elif opt=='cvboiyes':
    PrintChoice(['Digital Image Processing','Biomedical Image processing','Computer Vision','Advance Computer Vision',
           'Special Topics in Computer Vision'])
  elif opt=='cvboino':
    PrintChoice(['Digital Image Processing','Computer Vision','Advance Computer Vision','Special Topics in Computer Vision'])
  elif opt=='nlp':
    PrintChoice(['Natural Language Processing','Information Retrieval','Mining Large Networks'])
  elif opt=='sml':
    PrintChoice(['Statistical Machine Learning','Theories of Deep Learning','Theoretical Machine Learning,Bayesian Machine Learning',
           'Statistical Inference'])
  elif opt=='dbms':
    PrintChoice(['Fundamentals of Database Management System','Database System Implementation','Big Data Analytics'])
  elif opt=='is':
    PrintChoice(['Secure Coding','Theory of Modern Cryptography','Foundations to Computer Security',
           'Applied Cryptography','Network Security','Security Engineering','Internet Security & Privacy'])
  elif opt=='mc':
    PrintChoice(['Mobile Computing','Embedded Systems','Ad Hoc Wireless Networks','Cellular Data Networks',
           'Introduction to Spatial Computing','Smart Sensing for Internet of Things',
           'Advanced Topics in Mobile Computing'])      

def InputInterest():
  f = open("/content/drive/MyDrive/Projects/AI_Assignments/Assignment_5/userinput.txt", 'w')
  approach = {'1':'practical','2':'theory'}
  ft = {'1':'artificalintelligence','2':'database','3':'infosec','4':'mobilecomp'}
  SubAI = {'1':'robotics','2':'computervision','3':'machinelearning','4':'naturallanguage','5':'statisticalml'}
  IntBio = {'1':'yes','2':'no'}
  Prob = {'1':'yes','2':'no'}
  Exp = {'1':'less2','2':'more2'}
  domain = {'1':'algo','2':'socialscience','3':'compbiology','4':'maths'}

  print('Enter options number corresponding to given choices')

  app = input('\nYour approach :\n 1. Practical \n 2. Theoretical\n =>')
  f.write("approach("+approach[app]+").\n")

  if app=="1":
    fi = input('\nField of Technology :\n 1. Artifical Intelligence \n 2. DBMS \n 3. Information Security \n 4. Mobile Computing \n=>')
    f.write("domain("+ft[fi]+").\n")
    if fi=="1":
      sub = input('\nSubdomain in AI :\n 1. Robotics \n 2. Computer Vision \n 3. Machine Learning  \n 4. Natural Language Processing  \n 5. Statistical Aspects of Machine Learning \n=>')
      f.write("aidomain("+SubAI[sub]+").\n")
      if sub=="2":
        intb = input('\nInterest in Biology :\n 1. Yes \n 2. No \n=>')
        f.write("bio("+IntBio[intb]+").\n")
      if sub=="3":
        mlexp = input('\nExperience in Machine Learning :\n 1. Less than 2 years \n 2. More than 2 years \n=>')
        f.write("exper("+Exp[mlexp]+").\n")
        if mlexp=="1":
          probe = input('\nExperience in Probability and Statistics :\n 1. Yes \n 2. No \n=>')
          f.write("probability("+Prob[probe]+").\n")
          if probe=="1":
            intb = input('\nInterest in Biology :\n 1. Yes \n 2. No \n=>')
            f.write("bio("+IntBio[intb]+").\n")
  if app =="2":
    dom = input("\nDomain :\n 1. Algorithms \n 2. Social Science and Humanities \n 3. Biology \n 4. Mathematics \n=>")
    f.write("theorydomain("+domain[dom]+").\n")
  f.close()
    
```
