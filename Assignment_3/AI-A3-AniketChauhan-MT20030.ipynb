# For Courses Suggestion

```python
def PrintChoices(data):
  for i in range(len(data)):
    print(i+1,'. ',data[i])
```

## Durable rules

```python
fact = []
sugg = {}
with ruleset('course'):
    
    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='Robo'))
    def hci(e):
        e.assert_fact({ 'subject': ['Probability and Random Process','Reinforcement Learning','Robotics','Social Robotics']})
   
    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='CV') & (m.IntBio=='Yes'))
    def hci(e):
        e.assert_fact({ 'subject': ['Digital Image Processing','Biomedical Image processing','Computer Vision','Advance Computer Vision',
           'Special Topics in Computer Vision']})

    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='CV') & (m.IntBio=='No'))
    def hci(e):
        e.assert_fact({ 'subject': ['Digital Image Processing','Computer Vision','Advance Computer Vision','Special Topics in Computer Vision']})

    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='ML') & (m.Exp=='Lesser than 2 years') & (m.Prob =='Yes') & (m.IntBio=='Yes'))
    def hci(e):
        e.assert_fact({ 'subject': ['Machine Learning','Artifical Intelligence' ,'Data Mining','Data Science','Data Warehouse','Data Lifecycle Management',
           'Algorithms in Bioinformatics','Data Sciences for Genomics','Machine Learning for Biomedical Applications',
           'Biostatistics','Advanced Biometrics','Computing for Medicine']})

    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='ML') & (m.Exp=='Lesser than 2 years') & (m.Prob =='Yes') & (m.IntBio=='No'))
    def hci(e):
        e.assert_fact({ 'subject': ['Probability and Random Process','Probability and Statistics','Measure and Probability Theory',
           'Machine Learning','Data Mining']})

    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='ML') & (m.Exp=='Lesser than 2 years') & (m.Prob =='No'))
    def hci(e):
        e.assert_fact({ 'subject': ['Probability and Random Process','Probability and Statistics','Measure and Probability Theory',
           'Machine Learning','Data Mining']})
    
    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='ML') & (m.Exp=='More than 2 years'))
    def hci(e):
        e.assert_fact({ 'subject': ['Advance Machine Learning','Deep Learning','Probabilitic Graphical Model','Bayesian Machine Learning']})

    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='NLP'))
    def hci(e):
        e.assert_fact({ 'subject': ['Natural Language Processing','Information Retrieval','Mining Large Networks']})    

    @when_all((m.approach == 'Pratical') & (m.ft == 'AI') & (m.SubAI=='SML'))
    def hci(e):
        e.assert_fact({ 'subject': ['Statistical Machine Learning','Theories of Deep Learning','Theoretical Machine Learning,Bayesian Machine Learning',
           'Statistical Inference']})    
    
    @when_all((m.approach == 'Pratical') & (m.ft == 'DBMS'))
    def hci(e):
        e.assert_fact({ 'subject': ['Fundamentals of Database Management System','Database System Implementation','Big Data Analytics']})
    
    @when_all((m.approach == 'Pratical') & (m.ft == 'IS'))
    def hci(e):
        e.assert_fact({ 'subject': ['Secure Coding','Theory of Modern Cryptography','Foundations to Computer Security',
           'Applied Cryptography','Network Security','Security Engineering','Internet Security & Privacy']})
    
    @when_all((m.approach == 'Pratical') & (m.ft == 'MC'))
    def hci(e):
        e.assert_fact({ 'subject': ['Mobile Computing','Embedded Systems','Ad Hoc Wireless Networks','Cellular Data Networks',
           'Introduction to Spatial Computing','Smart Sensing for Internet of Things',
           'Advanced Topics in Mobile Computing']})
    
    @when_all((m.approach == 'Theoretical') & (m.domain == 'Algo'))
    def hci(e):
        e.assert_fact({ 'subject': ['Randomized Algorithms','Introduction to Graduate Algorithms','Approximation Algorithms',
           'Advanced Algorithms','Modern Algorithm Design']})
    
    @when_all((m.approach == 'Theoretical') & (m.domain == 'SSH'))
    def hci(e):
        e.assert_fact({ 'subject': ['Philosophy of Technology','Urban Space and Political Power','Comparative Politics in Digital Era',
           'Technology and the Future of Work','Advanced Sociological Theory','Advance Ethnographic Research Methods']})
    
    @when_all((m.approach == 'Theoretical') & (m.domain == 'Bio'))
    def hci(e):
        e.assert_fact({ 'subject': ['Foundations of Biology','Cell Biology and Bio-Chemistry','Cell Biology and Bio-Chemistry',
           'Foundations of Modern Biology','Introduction to Mathematical Biology','Introduction to Computational Neuroscience']})
    
    @when_all((m.approach == 'Theoretical') & (m.domain == 'Maths'))
    def hci(e):
        e.assert_fact({ 'subject': ['Linear Optimization','Advanced Linear Algebra','Algebraic Number Theory','Variational Calculus and their Applications',
           'Convex Optimization','Combinatorial Optimization']})
    
    @when_all(+m.subject)
    def output(c):
        fact.append(c.m.subject)
```

## User input 

```python
def InputInterest():
  
  approach = {'1':'Pratical','2':'Theoretical'}
  ft = {'1':'AI','2':'DBMS','3':'IS','4':'MC'}
  SubAI = {'1':'Robo','2':'CV','3':'ML','4':'NLP','5':'SML'}
  IntBio = {'1':'Yes','2':'No'}
  Prob = {'1':'Yes','2':'No'}
  Exp = {'1':'Lesser than 2 years','2':'More than 2 years'}
  domain = {'1':'Algo','2':'SSH','3':'Bio','4':'Maths'}

  print('Enter options number corresponding to given choices')
  inputdict = OrderedDict()
  inputtup = []

  app = input('\nYour approach :\n 1. Practical \n 2. Theoretical\n =>')
  inputdict['approach'] = approach[app]
  inputtup.append(approach[app])
  if app=="1":
    fi = input('\nField of Technology :\n 1. Artifical Intelligence \n 2. DBMS \n 3. Information Security \n 4. Mobile Computing \n=>')
    inputdict['ft'] = ft[fi]
    inputtup.append(ft[fi])
    if fi=="1":
      sub = input('\nSubdomain in AI :\n 1. Robotics \n 2. Computer Vision \n 3. Machine Learning  \n 4. Natural Language Processing  \n 5. Statistical Aspects of Machine Learning \n=>')
      inputdict['SubAI'] = SubAI[sub]
      inputtup.append(SubAI[sub])
      if sub=="2":
        intb = input('\nInterest in Biology :\n 1. Yes \n 2. No \n=>')
        inputdict['IntBio'] = IntBio[intb]
        inputtup.append(IntBio[intb])
      if sub=="3":
        mlexp = input('\nExperience in Machine Learning :\n 1. Less than 2 years \n 2. More than 2 years \n=>')
        inputdict['Exp'] = Exp[mlexp]
        inputtup.append(Exp[mlexp])
        if mlexp=="1":
          probe = input('\nExperience in Probability and Statistics :\n 1. Yes \n 2. No \n=>')
          inputdict['Prob'] = Prob[probe]
          inputtup.append(Prob[probe])
          if probe=="1":
            intb = input('\nInterest in Biology :\n 1. Yes \n 2. No \n=>')
            inputdict['IntBio'] = IntBio[intb]
            inputtup.append(IntBio[intb])
  if app =="2":
    dom = input("\nDomain :\n 1. Algorithms \n 2. Social Science and Humanities \n 3. Biology \n 4. Mathematics \n=>")
    inputdict['domain'] = domain[dom]     
    inputtup.append(domain[dom])   

  return dict(inputdict) , tuple(inputtup)


```

## Forward Chaining and Electives Suggestion

```python
inputdict , inputtup = InputInterest()
print()
try :
  assert_fact('course',inputdict) 
  sugg[inputtup] = fact[len(fact)-1]
  print('You can opt following courses : \n')
  PrintChoices(sugg[inputtup])
except :  
  print('You can opt following courses : \n')
  PrintChoices(sugg[inputtup])
```

# For Extracurricular Activities Suggestion

## Durable rules

```python
factext = []
suggext = {}
with ruleset('extracurricular'):
    
    @when_all((m.cate == 'academic') & (m.subcate == 'writing'))
    def hci(e):
        e.assert_fact({ 'subject': ['Participate in Essay writing','Story Writing ect.','Other Competitions related to writing']})
   
    @when_all((m.cate == 'academic') & (m.subcate == 'tech'))
    def hci(e):
        e.assert_fact({ 'subject': ['Participate in Coding Competitions','Join coding clubs and organise the coding competitions']})

    @when_all((m.cate == 'employ'))
    def hci(e):
        e.assert_fact({ 'subject': ['Participate in election for committee Positions' ,  'Pitch the ideas for startup','Organsise fund raising , etc.']})

    @when_all((m.cate == 'volt'))
    def hci(e):
        e.assert_fact({ 'subject': ['Participate in Public speaking','Debate competition' ,  'Organsie the talks pertaining to specific topic , etc.']})

    @when_all((m.cate == 'games') & (m.gtype =='indoor'))
    def hci(e):
        e.assert_fact({ 'subject': ['Card Games','Table Tennis','Chess']})

    @when_all((m.cate == 'games') & (m.gtype =='outdoor'))
    def hci(e):
        e.assert_fact({ 'subject': ['Cricket','Football','Martial Arts','Race,etc.']})
            
    @when_all((m.cate == 'cult'))
    def hci(e):
        e.assert_fact({ 'subject': ['Dancing','Singing','Drama Play,etc.']})    
        
    @when_all(+m.subject)
    def output(c):
        factext.append(c.m.subject)
```

## User Input

```python
def InputCate():
  
  cate = {'1':'academic','2':'employ','3':'volt','4':'games','5':'cult'}
  subcate = {'1':'writing','2':'tech'}
  gtype = {'1':'indoor','2':'outdoor'}

  print('Enter options number corresponding to given choices')
  inputdict = OrderedDict()
  inputtup = []

  app = input('\nYour approach :\n 1. Academic \n 2. Employment \n 3. Volunteering \n 4. Games \n 5. Cultural \n =>')
  inputdict['cate'] = cate[app]
  inputtup.append(cate[app])
  if app=="1":
    fi = input('\nInterest Subcategory :\n 1. Writing \n 2. Technology \n=>')
    inputdict['subcate'] = subcate[fi]
    inputtup.append(subcate[fi])
  if app =="4":
    dom = input("\nType :\n 1. Indoor \n 2. Outdoor \n=>")
    inputdict['gtype'] = gtype[dom]     
    inputtup.append(gtype[dom])   

  return dict(inputdict) , tuple(inputtup)
```

## Forward Chaining and Extracurricular Activites Suggestion

```python
inputdict , inputtup = InputCate()
print()
try :
  assert_fact('extracurricular',inputdict) 
  suggext[inputtup] = factext[len(factext)-1]
  print('You can go for : \n')
  PrintChoices(suggext[inputtup])
except :  
  print('You can go for : \n')
  PrintChoices(suggext[inputtup])
```
